```python
import pandas as pd
import numpy as np
import seaborn as sb


movies = pd.read_csv('movies_metadata.csv', low_memory= False, skiprows=[6362])
rating = pd.read_csv('ratings.csv')

movies= movies[['id', 'original_title', 'original_language', 'vote_count']]

movies.rename(columns= {'id' : 'MOVIE_ID', 'original_title':'TITLE', 'original_language': 'LANGUAGE', 'vote_count':'VOTE_COUNT'}, inplace= True)

movies.head()
```

```python
rating = rating [['userId', 'movieId', 'rating']]
rating.rename(columns={'userId': 'USER_ID', 'movieId': 'MOVIE_ID', 'rating': 'RATING'}, inplace= True)
rating.head()
```

```python
rating['USER_ID'].value_counts()
```

```python
count_rating= rating['USER_ID'].value_counts()>999
y= count_rating[count_rating].index
y.shape
```

```python
rating= rating[rating['USER_ID'].isin(y)]
rating.shape
```

```python
movies = movies[movies['VOTE_COUNT']>999]
movies['LANGUAGE'].value_counts().head(20)
```

```python
movies.info()
rating.info()
```

```python
movies_and_rating= rating.merge(movies, on='MOVIE_ID')
movies_and_rating.shape
```

```python
movies_and_rating.drop_duplicates(['USER_ID', 'MOVIE_ID'], inplace=True)
movies_and_rating.shape
```

```python
del movies_and_rating['MOVIE_ID']
```

```python
movies_and_rating.head()
```

```python
movie_pivot= movies_and_rating.pivot_table(columns= 'USER_ID', index= 'TITLE', values='VOTE_COUNT')
movie_pivot.fillna(0, inplace=True)
movie_pivot.head(20)
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

movies_sparse= csr_matrix(movie_pivot)
type(movies_sparse)
model= NearestNeighbors(algorithm= 'brute')
model.fit(movies_sparse)
```

```python
distances, sugestions = model.kneighbors(movie_pivot.filter(items= ['8 Mile'], axis=0).values.reshape(1, -1))

for i in range(len(sugestions)):
    print(movie_pivot.index[sugestions[i]])
```
